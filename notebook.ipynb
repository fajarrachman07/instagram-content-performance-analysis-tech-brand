- Perbedaan antara nilai maksimum engagement rate (22,34%) dan kuartil atas (5,66%) mengindikasikan adanya outlier, yaitu sejumlah kecil konten dengan performa engagement yang sangat tinggi.
- Untuk metrik reach, rata-rata berada pada angka 6.092, sementara median hanya 4.799, yang menunjukkan distribusi condong ke kanan (right-skewed) akibat beberapa konten dengan jangkauan sangat besar.
- Nilai maksimum reach mencapai 50.907, jauh di atas nilai kuartil atas (7.727), yang mengonfirmasi keberadaan konten dengan performa luar biasa dalam menjangkau audiens.
- Standar deviasi yang relatif besar pada kedua metrik menunjukkan variasi performa yang tinggi, di mana sebagian kecil konten berkontribusi signifikan terhadap rata-rata keseluruhan.

Kesimpulan deskriptif:
Distribusi engagement rate dan reach pada konten Instagram brand teknologi tidak merata. Sebagian kecil konten dengan performa sangat tinggi berperan besar dalam menaikkan nilai rata-rata, sehingga analisis lanjutan diperlukan untuk memahami faktor-faktor yang membedakan konten berperforma tinggi tersebut.

```python
# Uji Perbedaan Engagement Rate antar Media Type (ANOVA)

# Menjawab penjabaran masalah inferensial (6):
# Apakah terdapat perbedaan engagement rate yang signifikan antar media type?

from scipy.stats import f_oneway

reel_er = df_clean[df_clean["media_type"] == "reel"]["engagement_rate_pct"]
image_er = df_clean[df_clean["media_type"] == "image"]["engagement_rate_pct"]
carousel_er = df_clean[df_clean["media_type"] == "carousel"]["engagement_rate_pct"]

anova_result = f_oneway(reel_er, image_er, carousel_er)
anova_result
```

**Insight:**

- Hasil uji ANOVA satu arah menunjukkan nilai F-statistic sebesar 0,14 dengan p-value sebesar 0,87.
- Karena nilai p-value > 0,05, maka tidak terdapat perbedaan engagement rate yang signifikan secara statistik   antara jenis media konten reel, image, dan carousel pada konten Instagram brand teknologi.
- Dengan demikian, hipotesis nol (H0) yang menyatakan bahwa rata-rata engagement rate antar media type adalah sama tidak dapat ditolak.

Kesimpulan inferensial:
Jenis media konten tidak memiliki pengaruh yang signifikan terhadap engagement rate. Hal ini mengindikasikan bahwa faktor lain di luar format media—seperti kualitas konten, pesan, dan relevansi informasi—kemungkinan lebih berperan dalam mendorong engagement pada konten Instagram brand teknologi.

Hal ini mengindikasikan bahwa pemilihan format media saja belum cukup untuk meningkatkan engagement secara konsisten.

Konten yang masuk kategori **Viral** memiliki karakteristik utama berupa **jangkauan yang sangat tinggi** serta **interaksi yang didominasi oleh saves**, dibandingkan shares dan comments. Pola ini menunjukkan bahwa konten teknologi yang viral cenderung bersifat **informatif, relevan, dan bernilai guna**, sehingga mendorong audiens untuk menyimpan konten sebagai referensi.

Dari sisi distribusi format, konten viral tidak hanya didominasi oleh video, tetapi juga banyak berasal dari **image dan carousel**, yang menegaskan bahwa konten visual statis tetap memiliki potensi besar untuk mencapai performa tinggi apabila dikemas dengan pesan yang tepat.

- Mengoptimalkan konten **image dan carousel** yang bersifat edukatif dan mudah dipahami.
- Merancang konten yang mendorong audiens untuk **menyimpan dan membagikan** informasi, karena saves terbukti
menjadi indikator kuat konten viral.
